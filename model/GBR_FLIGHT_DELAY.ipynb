```python
cat_col=new_air[['AIRLINE', 'ORIGIN_AIRPORT', 'DESTINATION_AIRPORT']]
cat_col
```

```python
df_cat=pd.get_dummies(cat_col,drop_first=True,prefix='', prefix_sep='')
```

```python
df=new_air[['MONTH', 'DAY', 'DAY_OF_WEEK','SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 
            'DEPARTURE_DELAY', 'TAXI_OUT','SCHEDULED_ARRIVAL','ARRIVAL_DELAY']]
```

```python
df=pd.concat([df,df_cat],axis=1)
df.head()
```

```python
df_sam=df.sample(n=100000)
df_sam.reset_index(drop=True)
```

```python
X_1=df_sam.drop(["ARRIVAL_DELAY"],axis=1)
Y_1=df_sam.ARRIVAL_DELAY
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
X1_train, X1_test, y1_train, y1_test = train_test_split(X_1, Y_1, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbc = GradientBoostingRegressor(n_estimators=20,random_state=0)
gbc.fit(X1_train,y1_train)
```

```python
print(gbc.score(X1_test,y1_test))
```

```python
print(gbc.score(X1_train,y1_train))
```

```python
y1_pred = gbc.predict(X1_test)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y1_test,  y1_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y1_test,   y1_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y1_test,   y1_pred)))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

def rmse(a, b):
    return np.sqrt(mean_squared_error(a, b))
```

```python
def test_params(**params):
    model = GradientBoostingRegressor( random_state=42, **params)
    model.fit(X1_train, y1_train)
    prd= model.predict(X1_test)
    train_rmse = rmse(model.predict(X1_train), y1_train)
    test_rmse = rmse(model.predict(X1_test), y1_test)
    Rmse = rmse(y1_test,prd)
    print('Train RMSE: {}, Test RMSE: {},RMSE:{}'.format(train_rmse, test_rmse,Rmse))
```

```python
def predict_price(AIRLINE,ORIGIN_AIRPORT, DESTINATION_AIRPORT,MONTH, DAY, DAY_OF_WEEK,SCHEDULED_DEPARTURE, DEPARTURE_TIME, 
            DEPARTURE_DELAY, TAXI_OUT,SCHEDULED_ARRIVAL):    
    loc_index = np.where(X_1.columns==AIRLINE)[0][0]
    loc1_index = np.where(X_1.columns==ORIGIN_AIRPORT)[0][0]
    loc2_index = np.where(X_1.columns==DESTINATION_AIRPORT)[0][0]

    x = np.zeros(len(X_1.columns))
    x[0] = MONTH
    x[1] = DAY
    x[2] = DAY_OF_WEEK
    x[3] = SCHEDULED_DEPARTURE
    x[4] = DEPARTURE_TIME
    x[5] = DEPARTURE_DELAY
    x[6] = TAXI_OUT
    x[7] = SCHEDULED_ARRIVAL
    
    if loc_index >= 0:
        x[loc_index] = 1
    if loc1_index >= 0:
        x[loc1_index] = 1
    if loc2_index >= 0:
        x[loc2_index] = 1

    return modelhy.predict([x])[0]
```
